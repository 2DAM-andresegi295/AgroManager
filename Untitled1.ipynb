**3001 - Galleta**

Definir una clase Galleta que tiene dos atributos: nombre de la
galleta y su forma. La clase Galleta debe de contener el
contructor y el siguiente método:

o __init__(self,nombre): un contructor que permite
inicializar los atributos de la clase Galleta.

o __hornear__(self): un método que permite mostrar dos
frases en la consola:

  ✔ Esta galleta con chispas de chocolate ha sido horneada en forma de estrella
✔ Buen provecho.


```python
class Galleta:
    def __init__(self, nombre, forma):
        self.nombre = nombre
        self.forma = forma

    def __hornear__(self):
        print(f'Esta {self.nombre} ha sido horneada con {self.forma}')
        print('Buen provecho')
galleta= Galleta('galleta de chocolate','avión')
galleta.__hornear__()
```

**3002 - Libro**

Definir una clase Libro que tienes 3 atributos, título del
libro, nombre del autor y precio del libro.
La clase Libro debe contener el constructor y el siguiente
método:

**o __init__(self,titulo,autor,precio):** constructor que
permite inicializar los atributos de la clase Libro

**o Mostrar_informaciones(self):** permite mostrar la
información relacionada con el libro.


```python
class Libro :
    def __init__(self,titulos,autor,precio):
        self.titulos = titulos
        self.autor = autor
        self.precio = precio

    def Mostrar_informaciones(self):
        print(f'El libro {self.titulos} fue escrito por { self.autor} y tiene un precio de { self.precio} €')

libro=Libro('El mendigo','jack','12.00')
libro.Mostrar_informaciones()
```

3003 - Nota

**texto en negrita**Definir una clase Nota que tiene 2 atributos, nombre del
estudiante y su nota. La clase Nota debe contener el constructor
y el siguiente método:

  **o __init__(self,nota,nombreEstudiante):** constructor que
  permite inicializar los atributos de la clase Nota
  **o ha_pasado(self):**
  El método muestra en consola si el alumno
  está aprobado o suspenso. Si la nota del estudiante supera
  los 75 puntos entonces el alumno ha aprobado, de lo
  contrario está suspenso.


```python
class Nota:
    def __init__(self,nota,nombreEstudiante):
        self.nota = nota
        self.nombreEstudiante = nombreEstudiante
    def ha_pasado(self):
        if self.nota>75:
            print(f'{self.nombreEstudiante} ha aprobado con un {self.nota}')
        else:
            print(f'{self.nombreEstudiante} ha suspendido con un {self.nota}')

nota =Nota(78,'Andrés')
nota.ha_pasado()
nota =Nota(59,'Andrés')
nota.ha_pasado()
```

**3004 - Círculo**

Definir una clase Circulo que tiene 3 atributos: el radio y la
posición de su centro O(x,y). La clase debe de contener el
constructor y los siguientes métodos:

**o __init__(self,x,y,r):**
constructor que permite inicializar
los atributos de la clase Circulo
**o area(self):**

calcula el área y la muestra
o perimetro(self): calcula el perímetro y lo muestra
**o mostrar_propiedades(self):**

muestra los valores de los
atributos por consola.

```python
import math

class Circulo:
    def __init__(self,x,y,r):
        self.x = x
        self.y = y
        self.r = r

    def area(self):
        print(f'El area de este círculo es {math.pi*(self.r*self.r)}')
    def perimetro(self):
        print(f'El perímetro de este círculo es {2*math.pi*2}')
    def mostrar_propiedades(self):
        print(f'X: {self.x}, Y: {self.y}, R: {self.r}')
circulo =Circulo(50,50,50)
circulo.area()
circulo.perimetro()
circulo.mostrar_propiedades()
```
