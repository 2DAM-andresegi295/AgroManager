**3004 - Círculo**

Definir una clase Circulo que tiene 3 atributos: el radio y la
posición de su centro O(x,y). La clase debe de contener el
constructor y los siguientes métodos:

**o __init__(self,x,y,r):**

los atributos de la clase Circulo

**3005 - Operación**

Definir una clase Operacion que tiene dos atributos: dos números
enteros x e y. Esta clase nos permitirá realizar operaciones
aritméticas entre con estos dos atributos. La clase Operación
debe contener el constructor y los siguientes métodos:

**o __init__(self,x,y):**
 constructor que permite inicializar
los atributos de la clase Operacion

**o suma(self):**

realiza y muestra la suma de los atributos

**o multiplicación(self):**
 realiza y muestra la multiplicación
o division(self): realiza y muestra la división. Debe
controla que no se puede dividir entre 0.

```python
class Operacion:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def suma(self):
        resultado = self.x + self.y
        print(f"La suma de {self.x} y {self.y} es: {resultado}")

    def multiplicacion(self):
        resultado = self.x * self.y
        print(f"La multiplicación de {self.x} por {self.y} es: {resultado}")

    def division(self):
        if self.y == 0:
            print("Error: No se puede dividir entre 0.")
        else:
            resultado = self.x / self.y
            print(f"La división de {self.x} entre {self.y} es: {resultado}")

op = Operacion(10, 5)
op.suma()
op.multiplicacion()
op.division()
op_cero = Operacion(8, 0)
op_cero.division()
```

3009 - Vehículo
Definir una clase Vehiculo y una clase Coche que herede de la
clase Vehiculo. La clase Vehiculo tiene 2 atributos: marca y
velocidad inicial, con un valor predeterminado = 0.
La subclase Coche tiene un atributo adicional llamado bocina con
un valor predeterminado de “tuuut”.
La clase Vehiculo debe contener el constructor y los siguientes
métodos:
o __init__(self,…): constructor que permite inicializar
atributos de la clase Vehiculo.
o acelerar(self,v): aumenta en v la velocidad
o desacelerar(self,v): disminuye en v la velocidad
o mostrar_velocidad(self): muestra la velocidad del vehículo
La subclase Coche debe contener el constructor y el siguiente
método adicional:
o __init__(self,…): constructor que permite inicializar los
atributos.
o Tocar_claxon(self): muestra por consola una cadena de
caracteres que expresa el sonido de la bocina.

```python
class Vehiculo:
    def __init__(self, marca, velocidad_inicial):
        self.marca = marca
        self.velocidad_inicial = velocidad_inicial

    def acelerar(self,v):
        self.velocidad_inicial += v

    def desacelerar(self,v):
        self.velocidad_inicial -= v

    def mostrar_velocidad(self):
        print("Velocidad: ",self.velocidad_inicial)

class Coche(Vehiculo):
    def __init__(self, marca, velocidad_inicial):
        super().__init__(marca, velocidad_inicial)
        self.bocina='tuuut'

    def tocar_claxon(self):
        print("Claxon: ",self.bocina)

coche=Coche(marca='honda',velocidad_inicial=10)
coche.mostrar_velocidad()
coche.acelerar(5)
coche.mostrar_velocidad()
coche.desacelerar(5)
coche.mostrar_velocidad()
coche.tocar_claxon()
```

3010 - Empleado
Definir una clase Empleado con 4 atributos, el nombre, función,
salario y número de horas trabajadas.
La clase Empleado debe contener el constructor y los siguientes
métodos:
o __init__(self,nombre,función,salario): constructor que
permite inicializar los atributos. El número de horas
trabajadas se inicia a 0 en el constructor.
o trabajar(self,num_horas): aumenta en num_horas las horas
ya trabajadas. Además devuelve una cadena con el número de
horas trabajadas.
o info_sueldo(self): devuelve una cadena que muestra el
salario del empleado.
o dar_aumento(self,cantidad): incrementa en cantidad el
salario del empleado
o info_funcion(self): devuelve una cadena con la función del
empleado


```python
class Empleado:
    def __init__(self, nombre, funcion, salario):
        self.nombre = nombre
        self.funcion = funcion
        self.salario = salario
        self.horas_trabajadas = 0

    def trabajar(self,num_horas)->str:
        self.horas_trabajadas += num_horas
        return f"El empleado ha trabajado {self.horas_trabajadas} horas"

    def info_sueldo(self)->str:
        return f"El empleado cobra {self.salario}€"

    def dar_aumento(self,cantidad):
        self.salario += cantidad

    def info_funcion(self)->str:
        return f"El empleado tiene la función de {self.funcion}"

empleado = Empleado('Andrés', 'Desarrollo', 1500)
print(empleado.trabajar(40))
print(empleado.info_sueldo())
empleado.dar_aumento(500)
print(empleado.info_sueldo())
print(empleado.info_funcion())
```

3011 - Cálculo numérico
Definir una clase Calculo_numerico que permita llevar a cabo
varias operaciones numéricas. Esta clase tiene un solo atributo
que es un número entero. Debe contener el constructor y varios
métodos de operaciones numéricas:
o __init__(self numero): constructor que permite inicializar
el atributo de la clase.
o calculo_factorial(self): retorna el factorial del atributo
o lista_divisores(self): retorna una lista con los divisores
del atributo
o esPrimo(self): muestra por consola si el número es primo o
no.
o esPar(self): muestra por consola si el número es par o no.

```python


class CalculoNumerico:
    def __init__(self, numero:int):
        self.numero = numero
    def calculo_factorial(self)->int:
        n = self.numero
        if n < 0:
            raise ValueError("No existe factorial para números negativos")
        resultado = 1
        for i in range(2, n + 1):
            resultado *= i
        return resultado

    def lista_divisores(self) -> list[int]:
        n = self.numero
        if n == 0:
            raise ValueError("Los divisores de 0 no están definidos")
        n_abs = abs(n)
        divisores: list[int] = []
        i = 1
        while i * i <= n_abs:
            if n_abs % i == 0:
                divisores.append(i)
                otro = n_abs // i
                if otro != i:
                    divisores.append(otro)
            i += 1
        divisores.sort()
        return divisores

    def es_primo(self):
        n = self.numero
        if n <= 1:
            print(f"{n} no es primo")
            return
        limite = int(n ** 0.5)
        for i in range(2, limite + 1):
            if n % i == 0:
                print(f"{n} no es primo (divisible por {i})")
                return
        print(f"{n} es primo")

    def es_par(self):
        if self.numero % 2 == 0:
            print(f"{self.numero} es par")
        else:
            print(f"{self.numero} es impar")

calculo = CalculoNumerico(15)
print(calculo.calculo_factorial())
print(calculo.lista_divisores())
calculo.es_par()
calculo.es_primo()
```

3012 - Video - (Herencia múltiple)
Crear la clase Video y la clase Audio. Luego crea una subclase
Media que herede de las clases Video y Audio. Los atributos para
cada clase son los siguientes:
Video: título, duración en minutos y categoría
Audio: título, nombre del artista
Media: título, categoría, duración y nombre del artista
Cada clase tiene su constructor y los siguientes métodos:
Video:
o __init__(self, …): constructor que permite inicializar el
atributo de la clase Video.
o mirar_video(self): muestra por consola una cadena
indicando que estás viendo el video.
o detener_video(self): muestra por consola una cadena
indicando que has detenido la visualización del video.
Audio:
o __init__(self, …): constructor que permite inicializar el
atributo de la clase Audio.
o escuchar_audio(self): muestra por consola una cadena
indicando que estás escuchando el audio.
o detener_audio(self): muestra por consola una cadena
indicando que has detenido la audición del audio.
Media:
o __init__(self, …): constructor que permite inicializar el
atributo de la clase Media.

```python
class Video:
    def __init__(self, titulo, duracion, categoria):
        self.titulo = titulo
        self.duracion = duracion
        self.categoria = categoria

    def mirar_video(self):
        print("Estás viendo el video.")

    def detener_video(self):
        print("Has detenido la visualización del video.")

class Audio:
    def __init__(self, titulo, artista):
        self.titulo = titulo
        self.artista = artista

    def escuchar_audio(self):
        print("Estás escuchando el audio.")

    def detener_audio(self):
        print("Has detenido la audición del audio.")

class Media(Video, Audio):
    def __init__(self, titulo, categoria, duracion, artista):
        Video.__init__(self, titulo, duracion, categoria)
        Audio.__init__(self, titulo, artista)

media_player = Media("Bohemian Rhapsody", "Música", 6, "Queen")

media_player.mirar_video()
media_player.escuchar_audio()
media_player.detener_video()
media_player.detener_audio()
```
